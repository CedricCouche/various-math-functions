# Growth functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Logistic Growth

```python

def logistic_growth(t, L, k, t0):
    return L / (1 + np.exp(-k * (t - t0)))

# Example
t = np.linspace(0, 10, 100)
L, k, t0 = 100, 1, 5  # Parameters
y = logistic_growth(t, L, k, t0)

plt.plot(t, y, label="Logistic Growth")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()

plt.savefig("Logistic-Growth_20250130.png", transparent=None,format="png")
plt.show()
```

## Weibull Growth Function

```python
def weibull_growth(t, L, k, n):
    return L * (1 - np.exp(-(k * t) ** n))

# Example
t = np.linspace(0, 10, 100)
L, k, n = 100, 0.5, 2  # Parameters
y = weibull_growth(t, L, k, n)

plt.plot(t, y, label="Weibull Growth")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()

plt.savefig("Weibull_Growth_20250130.png", transparent=None,format="png")
plt.show()
```

## Monod Equation

```python
def monod_growth(t, mu_max, Ks):
    return (mu_max * t) / (Ks + t)

# Example
t = np.linspace(0, 10, 100)
mu_max, Ks = 100, 2  # Parameters
y = monod_growth(t, mu_max, Ks)

plt.plot(t, y, label="Monod Growth")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()

# plt.savefig("Monod_growth_20250130.png", transparent=None,format="png")
plt.show()
```

## Von Bertalanffy Growth Function

```python
def von_bertalanffy_growth(t, L, k, t0):
    return L * (1 - np.exp(-k * (t - t0))) ** 3

# Example
t = np.linspace(0, 10, 100)
L, k, t0 = 100, 1, 5  # Parameters
y = von_bertalanffy_growth(t, L, k, t0)

plt.plot(t, y, label="Von Bertalanffy Growth")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()

# plt.savefig("von_bertalanffy_growth_20250130.png", transparent=None,format="png")
plt.show()
```

## Exponential Decay to a Limit

```python
def exponential_decay_to_limit(t, L, k):
    return L * (1 - np.exp(-k * t))

# Example
t = np.linspace(0, 10, 100)
L, k = 100, 0.5  # Parameters
y = exponential_decay_to_limit(t, L, k)

plt.plot(t, y, label="Exponential Decay to Limit")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()

# plt.savefig("exponential_decay_to_limit_growth_20250130.png", transparent=None,format="png")
plt.show()
```

## Michaelis-Menten Equation

```python
def michaelis_menten(t, Vmax, Km):
    return (Vmax * t) / (Km + t)

# Example
t = np.linspace(0, 10, 100)
Vmax, Km = 100, 2  # Parameters
y = michaelis_menten(t, Vmax, Km)

plt.plot(t, y, label="Michaelis-Menten")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()

# plt.savefig("michaelis_menten_growth_20250130.png", transparent=None,format="png")
plt.show()
```

## Gompertz Growth Function

```python
def gompertz_growth(t, L, k, t0):
    return L * np.exp(-np.exp(-k * (t - t0)))

# Example
t = np.linspace(0, 10, 100)
L, k, t0 = 100, 1, 5  # Parameters
y = gompertz_growth(t, L, k, t0)

plt.plot(t, y, label="Gompertz Growth")
plt.xlabel("Time")
plt.ylabel("Growth")
plt.legend()

# plt.savefig("gompertz_growth_20250130.png", transparent=None,format="png")
plt.show()
```
