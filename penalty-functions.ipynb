# Penalty functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Hyperbolic Decay Function

```python
def hyperbolic_decay(t):
    return 1/(t+1)

# Example
t = np.linspace(0, 10, 100)
y = hyperbolic_decay(t)

plt.plot(t, y, label="hyperbolic decay")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# plt.savefig("hyperbolic_decay_20250130.png", transparent=None,format="png")
plt.show()
```

## Step Decay Function

```python
def step_decay(t, t_cutoff, c):
    return np.where(t<= t_cutoff, 1, c)

# Example
t = np.linspace(0, 36, 100)
t_cutoff = 12  # Treat the last 12 months equally
c = 0.1  # Penalty for values anterior to the last t_cutoff

y = step_decay(t, t_cutoff, c)

plt.plot(t, y, label="step decay")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# plt.savefig("step-decay_20250130.png", transparent=None,format="png")
plt.show()
```

## Logarithmic Decay Function

```python
def logarithmic_decay(t):
    return 1 / np.log(t + 2)


# Example
t = np.linspace(0, 10, 100)
y = logarithmic_decay(t)

plt.plot(t, y, label="logarithmic decay")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# plt.savefig("logarithmic_decay_20250130.png", transparent=None,format="png")
plt.show()
```

## Gaussian Decay Function


```python

def gaussian_decay(t, sigma):
    return np.exp(-(t** 2) / (2 * sigma ** 2))

# Example
t = np.linspace(0, 10, 100)

plt.plot(t, gaussian_decay(t, sigma=8), label="Gaussian decay with sigma = 8")
plt.plot(t, gaussian_decay(t, sigma=10), label="Gaussian decay with sigma = 10")
plt.plot(t, gaussian_decay(t, sigma=12), label="Gaussian decay with sigma = 12")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# plt.savefig("gaussian_decay_20250130.png", transparent=None,format="png")
plt.show()
```

## Linear Decay Function

## Power-Law Decay

```python
def power_law_decay(t, gamma):
    return  1 / (t + 1) ** gamma 

# Example
t = np.linspace(0, 10, 100)
gamma = 1.0  # Adjust gamma to control decay rate
y = power_law_decay(t, gamma)

plt.plot(t, y, label="Power Law decay")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# plt.savefig("power_law_decay_20250130.png", transparent=None,format="png")
plt.show()
```

## Exponential Decay

```python
def exponential_decay(t, lambda_):
    return np.exp(-lambda_ * t)


# Example
t = np.linspace(0, 10, 100)
lambda_ = 0.1  # Adjust lambda to control decay rate

y = exponential_decay(t, lambda_)

plt.plot(t, y, label="Exponential decay")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# plt.savefig("Exponential_decay_20250130.png", transparent=None,format="png")
plt.show()
```
